```python
#replacing null values with empty string fillna() 
news_dataset=news_dataset.fillna('') 
```

```python
#merging the author name and title
news_dataset['content']=news_dataset['author']+' '+news_dataset['title']
```

```python
#seperating data nd label
X=news_dataset.drop(columns='label',axis=1)
Y=news_dataset[['label']]
```

```python
print(X)
print(Y)
```

Stemming

```python
port_stem=PorterStemmer()
```

```python
#whole stemming process  
def stemming(content):
    stemmed_content=re.sub('[^a-zA-Z]',' ',content) #removing extra characters with ' ' excluding latters from atoz and AtoZ
    stemmed_content=stemmed_content.lower()
    stemmed_content=stemmed_content.split()
    stemmed_content=[port_stem.stem(word) for word in stemmed_content if not word in stopwords.words('english')]
    stemmed_content=' '.join(stemmed_content)
    return stemmed_content
```

```python
news_dataset['content']=news_dataset['content'].apply(stemming)
```

```python
print(news_dataset['content'])
```

```python
#seperating data and label
X=news_dataset['content'].values
Y=news_dataset['label'].values
```

```python
print(X)
```

```python
print(Y)
```

```python
Y.shape
```

Converting textual data to numerical data

```python
X=news_dataset['content']
vectorizer=TfidfVectorizer() #common words ki low score important words ki high score
vectorizer.fit(X)

X=vectorizer.transform(X) #Y ki avsaram le just real or fake no pattern finding required
```

```python
print(X)
```

Splitting dataset to training and test data

```python
X_train,X_test,Y_train,Y_test=train_test_split(X,Y,test_size=0.2,stratify=Y,random_state=2)
```

Training Logistic regression model

```python
model=LogisticRegression()
```

```python
model.fit(X_train,Y_train)
```

Finding accuracy score

```python
X_train_prediction=model.predict(X_train)
training_data_accuracy=accuracy_score(X_train_prediction,Y_train)
print('accuracy score of training data:',training_data_accuracy)
```

```python
X_test_prediction=model.predict(X_test)
test_data_accuracy=accuracy_score(X_test_prediction,Y_test)
print('accuracy score of testing data:',test_data_accuracy)
```

Making a predictive system

```python
X_new=X_test[0]
prediction=model.predict(X_new)
print(prediction)
```
